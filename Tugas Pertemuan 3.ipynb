# 1. Program Python Menghitung Luas dan Keliling Persegi Panjang

```python
# Meminta input dari pengguna
panjang = float(input("Masukkan panjang persegi panjang: "))
lebar = float(input("Masukkan lebar persegi panjang: "))

# Menghitung luas dan keliling
luas = panjang * lebar
keliling = 2 * (panjang + lebar)

# Menampilkan hasil
print(f"Luas persegi panjang adalah: {luas}")
print(f"Keliling persegi panjang adalah: {keliling}")
```

# 2. Program Python Mencari Nilai Maksimum dalam List

```python
# Meminta input list dari pengguna
data = input("Masukkan daftar angka (pisahkan dengan spasi): ")

# Mengubah string input menjadi list angka
angka_list = [float(x) for x in data.split()]

# Mencari nilai maksimum
maksimum = max(angka_list)

# Menampilkan hasil
print(f"Nilai maksimum dalam list adalah: {maksimum}")
```

# 3. Program Python dengan 4 Fungsi (max, min, odd, even)

```python
# Fungsi-fungsi yang diminta

def fungsi_max(a, b):
    return a if a > b else b

def fungsi_min(a, b):
    return a if a < b else b

def fungsi_odd(a, b):
    if a % 2 != 0:
        return a
    elif b % 2 != 0:
        return b
    else:
        return "Tidak ada bilangan ganjil"

def fungsi_even(a, b):
    if a % 2 == 0:
        return a
    elif b % 2 == 0:
        return b
    else:
        return "Tidak ada bilangan genap"

# Meminta input dari pengguna
angka1 = int(input("Masukkan angka pertama: "))
angka2 = int(input("Masukkan angka kedua: "))

# Menampilkan hasil dari masing-masing fungsi
print(f"Nilai maksimum dari {angka1} dan {angka2} adalah: {fungsi_max(angka1, angka2)}")
print(f"Nilai minimum dari {angka1} dan {angka2} adalah: {fungsi_min(angka1, angka2)}")
print(f"Bilangan ganjil dari {angka1} dan {angka2} adalah: {fungsi_odd(angka1, angka2)}")
print(f"Bilangan genap dari {angka1} dan {angka2} adalah: {fungsi_even(angka1, angka2)}")
```
